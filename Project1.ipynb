LOAN APPLICATION DATA ANALYSIS
The “Loan Application” dataset comprises information related to credit card applications, including the card type, issuing bank, card number, cardholder’s name, CVV/CVV2, issue date, expiry date, billing date, card PIN, and credit limit. This dataset is valuable for analyzing credit card usage patterns, assessing creditworthiness, and predicting loan approval outcomes. Researchers and financial institutions can leverage this data to develop credit risk models, improve loan approval processes, and enhance fraud detection mechanisms.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
github_url = ('https://raw.githubusercontent.com/Srinathp712/Loan/main/loan_data.csv')
df = pd.read_csv(github_url, encoding='Windows-1252')
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Issuing Bank': ['Bank X', 'Bank Y', 'Bank X', 'Bank Y', 'Bank Z'],
    'Card Type Full Name': ['Visa', 'Mastercard', 'Visa', 'Visa', 'Mastercard']
}
grouped_df = df.groupby(['Issuing Bank', 'Card Type Full Name']).size().unstack()
grouped_df.plot(kind='bar', stacked=True)
plt.xlabel('Issuing Bank')
plt.ylabel('Count')
plt.title('Count of Card Types by Issuing Bank')
plt.legend(title='Card Type')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Card Type Full Name': ['Visa', 'Mastercard', 'Visa', 'Visa', 'Mastercard'],
    'Credit Limit': [5000, 7000, 6000, 5500, 8000]
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Card Type Full Name', y='Credit Limit', data=df)
plt.title('Distribution of Credit Limits by Card Type (Box Plot)')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Credit Limit', hue='Card Type Full Name', element='step', kde=True)
plt.title('Distribution of Credit Limits by Card Type (Histogram)')
plt.xlabel('Credit Limit')
plt.ylabel('Frequency')
plt.legend(title='Card Type')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Expiry Date': ['2023-01', '2023-02', '2023-03', '2023-04', '2023-05'],
    'Card Type Full Name': ['Visa', 'Mastercard', 'Visa', 'Visa', 'Mastercard']
}

df = pd.DataFrame(data)
df['Expiry Date'] = pd.to_datetime(df['Expiry Date'])

df['Expiry Month'] = df['Expiry Date'].dt.month
df['Expiry Year'] = df['Expiry Date'].dt.year

count_by_month_year = df.groupby(['Expiry Year', 'Expiry Month']).size()

plt.figure(figsize=(10, 6))
count_by_month_year.plot(kind='line', marker='o')
plt.xlabel('Expiry Date')
plt.ylabel('Count')
plt.title('Count of Cards Expiring Each Month/Year')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Billing Frequency': ['Monthly', 'Quarterly', 'Annually', 'Monthly', 'Annually', 'Monthly'],
    'Card Type Full Name': ['Visa', 'Mastercard', 'Visa', 'Visa', 'Mastercard', 'Visa']
}

df = pd.DataFrame(data)

billing_counts = df['Billing Frequency'].value_counts()

plt.figure(figsize=(8, 6))
billing_counts.plot(kind='pie', autopct='%1.1f%%', labels=None, startangle=90)
plt.title('Percentage of Cards by Billing Frequency')
plt.legend(labels=billing_counts.index, loc='upper right')
plt.axis('equal')
plt.show()

plt.figure(figsize=(10, 6))
billing_counts.plot(kind='bar')
plt.xlabel('Billing Frequency')
plt.ylabel('Count')
plt.title('Count of Cards by Billing Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Issue Date': pd.date_range(start='2022-01-01', end='2022-12-31', freq='M'),
    'Expiry Date': pd.date_range(start='2022-01-01', end='2022-12-31', freq='M'),
    'Billing Date': pd.date_range(start='2022-01-01', end='2022-12-31', freq='M')
}

df = pd.DataFrame(data)

issued_counts = df['Issue Date'].dt.month.value_counts().sort_index()
expired_counts = df['Expiry Date'].dt.month.value_counts().sort_index()
billed_counts = df['Billing Date'].dt.month.value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(issued_counts.index, issued_counts.values, label='Cards Issued')
plt.plot(expired_counts.index, expired_counts.values, label='Cards Expired')
plt.plot(billed_counts.index, billed_counts.values, label='Cards Billed')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Count of Cards Issued/Expired/Billed Over Time')
plt.legend()
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Expiry Date': ['2023-01', '2023-02', '2023-03', '2023-04', '2023-05'],
    'Card Type Full Name': ['Visa', 'Mastercard', 'Visa', 'Visa', 'Mastercard']
}

df = pd.DataFrame(data)
df['Expiry Month'] = pd.to_datetime(df['Expiry Date']).dt.strftime('%Y-%m')

grouped_df = df.groupby(['Expiry Month', 'Card Type Full Name']).size().unstack()

plt.figure(figsize=(10, 6))
grouped_df.plot(kind='bar', stacked=True)
plt.xlabel('Expiry Month')
plt.ylabel('Count')
plt.title('Count of Card Types Grouped by Expiry Month')
plt.legend(title='Card Type')
plt.show()
```
